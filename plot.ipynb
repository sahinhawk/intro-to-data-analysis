Does sugar content influence calorie count?

To answer this, I will first plot the two variables against each other to see their gradient and then I will calculate the correlation score to see numerically how well they correlate.

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('cereals.csv')

plt.scatter(df['sugars'],df['calories'])
plt.xlabel('sugar')
plt.ylabel('calories')
plt.title('Sugar v Calories')
plt.show()
```

```python
df['sugars'] = pd.to_numeric(df['sugars'],errors='coerce')
df['calories'] = pd.to_numeric(df['calories'],errors='coerce')
corr = df['sugars'].corr(df['calories'])
print(f"correlation between sugar and calorie count is: {corr}")
```

As the scatterplot has a positive gradient and the correlation score is 0.57 which is closer to 1 than to 0 meaning a positive correlation, I can conclude that there is a positive correlation between the two variables and that calorie count increases with sugar levels.

What are the top 4 cereals for vitamin content?

To answer this, I will use a bar chart to see which cereals have the highest vitamin levels.

First, I will reduce the size of the input data as many of the cereals actually have no vitamin content at all.

```python
df_vitamins = df[df.vitamins != 0]
```

```python
bar_chart = df_vitamins.plot.bar(x='name',y='vitamins',fontsize='5.5',figsize=(10,5.5))
```

I can see that 'Just Right Crunchy Nuggets', 'Just Right Fruit and Nut', 'Product 19', 'Total Corn Flakes', 'Total Raisin Bran' and 'Total Whole Grain' all have the joint highest vitamin content of 100 each.

Does Sodium content positively or negatively affect overall ranking?

I will answer this by using a scatterplot and calculating the correlation score between variables.

```python
plt.scatter(df['sodium'],df['rating'])
plt.xlabel('sodium content')
plt.ylabel('overall ranking')
plt.title('sodium content v overall ranking')
plt.show()
```

Initial observations: Negative gradient showing negative correlation, a higher sodium content is correlated with a lower overall ranking. Surprising as sodium is used to improve taste similar to sugar.

```python
df['sodium'] = pd.to_numeric(df['sodium'],errors='coerce')
df['rating'] = pd.to_numeric(df['rating'],errors='coerce')
corr = df['sodium'].corr(df['rating'])
print(f"correlation between sodium content and overall ranking count is: {corr}")
```

A negative gradient scatterplot as well as a correlation score closer to -1 than to 0 shows me that overall ranking decreases as sodium content increased which is surprising as you'd expect cereals with more sodium to be rated higher due to it being used for flavouring to improve taste.
